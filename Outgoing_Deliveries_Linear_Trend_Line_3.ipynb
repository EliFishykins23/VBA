```python
# Install Pandas Library:

import pandas as pd

#Install Numpy Library:

import numpy as np

# Google Drive Library:

from google.colab import drive
drive.mount('/content/drive')

# Actual Path In Google Drive ie: /content/drive/....(File Path) :

file_path = '/content/drive/My Drive/Data_Anayltics/Simple Linear Regression_Outgoing Deliveries.xlsx'

# Read the excel
df = pd.read_excel(file_path)

# Visualization Plot Chart - Matplotlib
import matplotlib.pyplot as plt

# - - - - - - - - - - - - - - - - - - - - - - - - -

# For Linear Trend Line - 'Pull Your X and Y data'
x = df['Delivery_Time_Hr']
y = df['Distance (km)']

# Fit a Linear Trend Line: y = mx + b
m, b = np.polyfit(x, y, 1) # degree = '1' for Linear

# - - - - - - - - - - - - - - - - - - - - - - - - -

# Plot Trend Line

plt.figure(figsize=(12, 6))
plt.scatter(x, y, marker='o', color='royalblue', label='Data Points')
plt.plot(x, m*x + b, color='orange', linewidth=1, label='Linear Trend Line')

plt.title('Delivery_Time_Hr vs Distance (km) with Linear Fit')
plt.xlabel('Delivery_Time_Hr')
plt.ylabel('Distance (km)')
plt.grid(True)
plt.legend
plt.tight_layout()
plt.show()

# Simple Linear Regression Model
# Sample size [n = 86]
# This model showcases actual delivery times for sample shipments given the amount in distance from start to end.
# If slope = 0.0045, then for every additional km in distance, delivery time increases by 0.0045 hours (~0.27 minutes) (small fraction of a min.) (~16.2 secs. to be exact).

# (Question): Is it possible the 2 points at the top are creating significant outliers (as extreme values), causing the coefficient to become distored and weak in regards to a linear projection?


```
